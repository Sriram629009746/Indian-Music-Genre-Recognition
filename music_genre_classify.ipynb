```python
import json
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
import joblib
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# path to json file that stores MFCCs and genre labels for each processed segment
DATA_PATH = "data_IndianMusic.json"
#RESULT_PATH = "result_random_forest.json"

GENRE_LIST = ["bollypop",
        "carnatic",
        "ghazal",
        "sufi"]
```

```python
def load_data(data_path):

    with open(data_path, "r") as fp:
        data = json.load(fp)

    # convert lists to numpy arrays
    X = np.array(data["mfcc"])
    y = np.array(data["labels"])

    print("Data succesfully loaded!")

    return  X, y
```

```python
def evaluate_model(model, X, y):
    # define the evaluation procedure
    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=42)
    # evaluate the model and collect the results
    scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
    return scores
```

# Feature Engineering

```python
# load data
x, y = load_data(DATA_PATH)
print (x.shape, y.shape)
```

```python
x_flattened = x.reshape(x.shape[0], -1)
print (x_flattened.shape)
```

```python
# create train/test split
X_train, X_test, y_train, y_test = train_test_split(x_flattened, y, test_size=0.2, random_state=42)
```

```python
scaler = sklearn.preprocessing.StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
pca = PCA(n_components=0.9)  # Retain 90% of variance
X_train = pca.fit_transform(X_train_scaled)
X_test = pca.transform(X_test_scaled)
```

```python
print(X_train.shape)
```

```python
from sklearn.model_selection import RepeatedStratifiedKFold, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
import tensorflow as tf
```

# Model Selection

```python
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'SVM': SVC(),
    'Random Forest': RandomForestClassifier(),
    'KNN': KNeighborsClassifier(),
    'Naive Bayes': GaussianNB(),
    'Decision Tree': DecisionTreeClassifier()
}

# Define hyperparameters for each model
param_grids = {
    'Logistic Regression': {'C': [0.1, 1, 10], 'penalty': ['l1', 'l2']},
    'SVM': {'C': [0.1, 1, 10], 'gamma': [0.1, 1, 10], 'kernel': ['rbf', 'linear']},
    'Random Forest': {'n_estimators': [25, 50, 100, 200], 'max_depth': [None, 5, 10, 20]},
    'KNN': {'n_neighbors': [3, 5, 7]},
    'Naive Bayes': {},
    'Decision Tree': {'max_depth': [None, 5, 10, 20], 'min_samples_split': [2, 5, 10]}
}

# Define the cross-validation strategy
cv_strategy = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=42)
```

```python
# Perform grid search for each model
best_models = {}
for model_name, model in models.items():
    print(f"Performing GridSearchCV for {model_name}...")
    grid_search = GridSearchCV(estimator=model, param_grid=param_grids[model_name], cv=cv_strategy, scoring='accuracy', return_train_score=True)
    grid_search.fit(X_train, y_train)
    best_models[model_name] = grid_search.best_estimator_
    print(f"Best Hyperparameters for {model_name}: {grid_search.best_params_}")
    
    
    cv_results = grid_search.cv_results_
    best_index = grid_search.best_index_
    
    train_score_from_results = grid_search.cv_results_['mean_train_score'][best_index]
    print("Train accuracy:", train_score_from_results)
    
    # Print cross-validation scores
    print(f"Mean CV Accuracy for {model_name} with best fit: {cv_results['mean_test_score'][best_index]}")
    
    #print(f"Mean CV Accuracy for {model_name}: {cv_results['mean_test_score'].mean()}\n")
    print()
    # Print cross-validation scores
    #cv_scores = cross_val_score(grid_search.best_estimator_, x_train_pca, y_train, cv=cv_strategy)
    #print(f"Cross-Validation Scores for {model_name}: {cv_scores}")
    #print(f"Mean CV Accuracy for {model_name}: {cv_scores.mean()}\n")
```

```python
X_train_NN, X_val_NN, y_train_NN, y_val_NN = train_test_split(X_train, y_train, test_size=0.1, random_state=42)
```

```python
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier

# Define a function to create a neural network model
def create_nn_model(units=64, activation='relu', optimizer='adam'):
    model = Sequential([
        Dense(units, activation=activation, input_shape=(X_train.shape[1],)),
        Dense(units//2, activation=activation),
        Dense(units//4, activation=activation),
        Dense(4, activation='softmax')
    ])
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Wrap the Keras model so it can be used by scikit-learn
nn_model = KerasClassifier(build_fn=create_nn_model, epochs = 50, verbose=0)

# Define the hyperparameter grid
param_grid = {
    'units': [64, 128, 256, 512],
    'activation': ['relu', 'tanh'],
    'optimizer': ['adam', 'rmsprop'],
    'batch_size': [32, 64, 128]
    
}

# Perform grid search for the neural network
grid_search = GridSearchCV(estimator=nn_model, param_grid=param_grid, cv=3, scoring='accuracy', verbose=0)
grid_search.fit(X_train, y_train)

# Get the best neural network model
print("Best Hyperparameters:", grid_search.best_params_)
```

```python
print("Mean CV Score with the best hyperparameters: {}".format(grid_search.cv_results_['mean_test_score'].max()))
```

**Based on the above CV scores, MLP model seems to be the best choice.**

# Evaluation 

Training the NN model with the above obtained hyperparameters

```python
# Define a neural network model based on the above hyperparameters
nn_model = Sequential([
    Dense(512, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(256, activation='relu'),
    Dense(128, activation='relu'),
    Dense(4, activation='softmax')
])
```

```python
nn_model.summary()
```

```python
# Compile the model
nn_model.compile(optimizer='rmsprop', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
nn_model.fit(X_train, y_train, epochs=50, batch_size=128, verbose=0)
```

**NN Model Results**

```python
y_pred_probs = nn_model.predict(X_test)  # Get predicted probabilities for each class
y_pred = np.argmax(y_pred_probs, axis=1)  # Convert probabilities to predicted class labels

# 2. Calculate accuracy
accuracy = np.mean(y_pred == y_test)  # Compare predicted labels with true labels and calculate accuracy
print("Test Accuracy of NN Model:", accuracy)

cm = tf.math.confusion_matrix(labels=y_test, predictions=y_pred).numpy()

con_mat_norm = np.around(cm.astype('float') / cm.sum(axis=1)[:, np.newaxis], decimals=2)

con_mat_df = pd.DataFrame(con_mat_norm,
                      index=GENRE_LIST,
                      columns=GENRE_LIST)

sns.heatmap(con_mat_df, annot=True, cmap=plt.cm.Blues)
```
